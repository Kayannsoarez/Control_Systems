```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

## Simulação da propagação do COVID-19 em um campus universitário para determinar o efeito das medidas de distanciamento social.

### Modelo Comportamental SEIR
A fração da população Suscetível, Exposta, Infectada e Recuperada (SEIR) é dada por um modelo comportamental com quatro equações diferenciais.

### $\frac{ds}{dt} = -(1-u)\beta s i$

### $\frac{de}{dt} = (1-u)\beta s i - \alpha e$

### $\frac{di}{dt} = \alpha e  - \gamma i$

### $\frac{dr}{dt} = \gamma i$

Suscetível(s): fração da população que é suscetível ao vírus

Exposto (e): fração da população está infectada com o vírus, mas não transmite para outras pessoas

Infecciosa (i): fração da população que está infectada e pode infectar outras

Recuperado (r): fração da população recuperada da infecção e está imune a novas infecções

***Altere o distanciamento social (u=0 (nenhum), u=1 (isolamento total)) para alterar o resultado***


```python
# Distanciamento Social (0-1)
u = 0.2

t_incubation = 5.1 
t_infective = 3.3 
R0 = 2.4
# População na UFC
N = 38650 

# Número inicial de indivíduos infectados e recuperados
e0 = 1/N
i0 = 0.00
r0 = 0.00
s0 = 1 - e0 - i0 - r0
x0 = [s0,e0,i0,r0]

alpha = 1/t_incubation
gamma = 1/t_infective
beta = R0*gamma

def covid(x,t):
    s,e,i,r = x
    dx = np.zeros(4)
    dx[0] = -(1-u)*beta * s * i
    dx[1] = (1-u)*beta * s * i - alpha * e
    dx[2] = alpha * e - gamma * i
    dx[3] = gamma*i
    return dx

t = np.linspace(0, 200, 101)
x = odeint(covid,x0,t)
s = x[:,0]; e = x[:,1]; i = x[:,2]; r = x[:,3]

# Plotagem do Grafico
plt.figure(figsize=(8,5))

plt.subplot(2,1,1)
plt.title('Distanciamento Social = '+str(u*100)+'%')
plt.plot(t,s, color='blue', lw=3, label='Suscetível')
plt.plot(t,r, color='red',  lw=3, label='Recuperado')
plt.ylabel('Fração')
plt.legend()

plt.subplot(2,1,2)
plt.plot(t,i, color='orange', lw=3, label='Infeccioso')
plt.plot(t,e, color='purple', lw=3, label='Exposto')
plt.ylim(0, 0.2)
plt.xlabel('Tempo (dias)')
plt.ylabel('Fração')
plt.legend()

plt.show()
```
